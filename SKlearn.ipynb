```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib as plt
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt2
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
%matplotlib inline

df = pd.read_csv("O:/Data Quality/Data Analytics/research/2020/6. Minerva/4. results/MIN101C07_101_mk_data_for_ML 12 April 2021.csv") #Reading the dataset in a dataframe using Pandas

df.head(10)

```

```python
tit = "Subset: AVAL_NSFS".title()
print(tit)
# subset
df=df[df["_NAME_"]=="AVAL_NSFS"]
display(df.head(10))


tit2 = "Numeric Variables Only".title()
print(tit2)
df=df.filter(regex='^V',axis=1)
display(df.head(10))
display(df.shape)

tit3 = "Exclude Missing".title()
print(tit3)
df=df.dropna()
display(df.head(10))
display(df.shape)

tit4 = "Standardized".title()
print(tit4)
#Standardize
clmns = ['V_1', 'V_3','V_7', 'V_9','V_10','V_11']
df2 = stats.zscore(df[clmns])
print(df2[1:10,])
display(df2.shape)
```

```python

#Cluster the data
km = KMeans(n_clusters=2, random_state=0).fit(df2)
labels = km.labels_
len(labels)
print(labels[1:10])
```

```python
### glue tabk to original data
#df3['clusters']=labels
df['clusters'] = labels
df.head(10)
```

```python
# add the column into list
clmns.extend(['clusters'])
print(df.columns)
```

```python

#Lets analyze the clusters
print (df.groupby('clusters').mean().reset_index())
```

```python
#filter rows of original data
filtered_label = df[df["clusters"]==1]
print(filtered_label.head(10))
#Plotting the results

## can't plot scatter plot for >2 dimentions unless two dimentioins are specified
plt2.scatter(filtered_label[:,0] , filtered_label[:,1] , color = 'red')

plt.show()
```
